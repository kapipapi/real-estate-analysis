```python
import os 

df1 = df1.drop(df1[df1['Numer oferty'] == '31093/3685/OMS'].index)

def count_photos(row):
    offer_id = row['Numer oferty']
    index = 0
    for root, _, files in os.walk(f"./images/{offer_id}"):
        for file in files:
            if file.endswith(".jpg") and "all" not in file and "wp" not in file:
                    index += 1

    if index == 0:
        print(f"No photos for offer {offer_id}")
        return None
    
    return row

df1.apply(count_photos, axis=1)
```
